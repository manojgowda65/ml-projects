## Named entity recognition
* each word is a one hot encoded with length = length of the dictionary of words
* We can't choose a traditional NN because of below reasons
1. Each training example(a sentence) will have varying length of words & each sentence might have different number of named entities in them. i.e there is no fixed inp/out size
2. As the NN's activations are independent and doesn't share features learned across other activations within a layer, with this approach we will not be able train a NLP model as we would require the model to share the features learned
3. Also, since each word is represented as one hot encoded vector of size N(length of dictionary), as the we dive deep into layers, each activation will enormous number of calculations to perform and there is a problem of memory

# Basic RNN's have vanishing gradient problem as the network is going to be huge. example with 10000 words dictinoary and a huge sentence being passed into model will have huge number of activations and hence vanishing gradient problem
* i.e simple RNN will fail to capture the initial activation feature maps that are required in the later layers to predict some other word
